プログラミングでは、日常生活のさまざまな状況を論理的に処理することができます。
この問題では、スーパーマーケットの割引システムをモデル化し、条件分岐（if文）を使って適切な割引率を決定する方法を学びます。

この問題を理解することで、以下のスキルを身につけることができます：

実際のビジネスロジックをプログラムで表現することで、プログラミングの実用的な側面を理解することができます。また、この問題は、より複雑な割引システムや価格計算のアルゴリズムへの入門としても役立ちます。

それでは、以下の問題文をよく読んで、アルゴリズム問題を理解しましょう。

### 問題文
あるスーパーマーケットでは、買い物の合計金額に応じて以下のような割引を行っています：

- 1000円未満：割引なし
- 1000円以上3000円未満：5%割引
- 3000円以上5000円未満：10%割引
- 5000円以上：15%割引

買い物の合計金額が与えられたとき、適用される割引率を求めるプログラムを作成してください。

### 制約
- 入力される金額は0以上10000以下の整数

##### 制約について
- 問題文中の「制約」は、あなたが作成するプログラムに課される制限ではありません。これは、入力として与えられる値の範囲や条件を示すものです。

##### 制約の意味
- 制約は、プログラムに入力される可能性のある値の範囲を定義します。
- 例えば、「Nは1以上100以下の整数」という制約は、入力値Nが1から100の間の整数であることを保証します。

##### なぜ制約が重要か
1. **エラー処理**: 制約外の入力を考慮する必要がないため、エラー処理を簡略化できます。
2. **最適化**: 入力値の範囲が分かることで、効率的なアルゴリズムを選択できます。
3. **テスト**: 制約に基づいてテストケースを作成し、プログラムの正確性を確認できます。

##### 制約の活用方法
- 制約を理解し、それに基づいてアルゴリズムを設計しましょう。
- 制約を満たす入力のみを想定してプログラムを作成します。
- 制約外の入力に対する特別な処理は不要です。

制約を正しく理解することで、より効率的で堅牢なプログラムを作成することができます。問題に取り組む際は、必ず制約をよく確認してください。

### 入力
入力は以下の形式で標準入力から与えられます。

---
```
N
```
---

- Nは買い物の合計金額を表す整数（0 ≤ N ≤ 10000）

### 出力
適用される割引率を整数で出力してください。単位（%）は不要です。

### 入力例1
---
```
2500
```
---

### 出力例1
---
```
5
```
---

### 入力例2
---
```
4999
```
---
    
### 出力例2
---
```
10
```
---

### 入力例3
---
```
7000
```
---

### 出力例3
---
```
15
```
---

### ここ↓にPythonコードを記述してください

````python
# 買い物金額から割引率を判定するプログラム

# 標準入力から買い物金額を受け取る
amount = int(input())

# 金額に応じて割引率を判定
if amount < 1000:
    discount = 0
elif amount < 3000:
    discount = 5
elif amount < 5000:
    discount = 10
else:
    discount = 15

# 結果を出力
print(discount)
````
