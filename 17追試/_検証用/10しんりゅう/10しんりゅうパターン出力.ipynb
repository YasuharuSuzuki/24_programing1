```python
def get_random_equipment():
    # 0: なし
    # 1: 珊瑚の指輪のみ
    # 2: ダイアの鎧のみ
    # 3: 両方
    pattern = random.randint(0, 3)
    
    if pattern == 0:
        return []
    elif pattern == 1:
        return ["珊瑚の指輪"]
    elif pattern == 2:
        return ["ダイアの鎧"]
    else:  # pattern == 3
        return ["珊瑚の指輪", "ダイアの鎧"]
```

```python
def get_random_character(force_heal=-1):
    hp_max = random.randint(CHARACTER_HP_MIN, CHARACTER_HP_MAX)
    equipment = get_random_equipment()
    wave = random.randint(WAVE_DAMAGE_MIN, WAVE_DAMAGE_MAX)
    thunder = random.randint(THUNDER_DAMAGE_MIN, THUNDER_DAMAGE_MAX)
    if force_heal > 0:
        heal = force_heal
    else:
        heal = random.randint(HEAL_MIN, HEAL_MAX)
    final_hp = calc_final_hp(hp_max, equipment, wave, thunder, heal)
    return hp_max, equipment, wave, thunder, heal, final_hp

get_random_character()
```

```python
# HP、装備、タイダルウエイブ、いなづま、回復量の組み合わせを作成。勝利 / 敗北を指定できます
def get_random_pattern(is_win_pattern, sample_count, force_heal=-1):
    dicts = []
    while len(dicts) < sample_count:
        hp_max, equipment, wave, thunder, force_heal, final_hp = get_random_character(force_heal)
        append_dict = {"hp_max":hp_max, "equipment":equipment, "wave":wave, "thunder":thunder, "heal":force_heal, "final_hp":final_hp}
        if is_win_pattern and final_hp > 0:
            dicts.append(append_dict)
        elif (not is_win_pattern) and final_hp < 1:
            dicts.append(append_dict)
    return pd.DataFrame(dicts)
get_random_pattern(False, 2, 3999)
```

```python
# キャラクタ単位のHP計算を行う
def calc_final_hp(hp_max, equipment, wave, thunder, heal):
    # タイダルウエイブの計算
    hp = hp_max
    damage = wave
    if "珊瑚の指輪" in equipment:
        damage //= 3
    hp -= damage
    print(f'タイダルウエイブ hp={hp}, damage={damage}')
    if hp<=0:
        return 0

    # 回復
    hp = min(hp_max, hp+heal)
    print(f'回復 hp={hp}, heal={heal}')

    # いなづまの計算
    damage = thunder
    if "珊瑚の指輪" in equipment:
        damage *= 2
    if "ダイアの鎧" in equipment:
        damage //= 3
    hp -= damage
    print(f'いなづま hp={hp}, damage={damage}')
    if hp<=0:
        return 0
    return hp
calc_final_hp(3667, ['珊瑚の指輪', 'ダイアの鎧'], 7999, 5499, 2667)
```

```python
# 自作のExcelデータからランダムで勝ちパターンのデータを取得する
def get_random_by_handmade(is_win_pattern, sample_count):

    if is_win_pattern:
        df_test_case_tmp1 = df_test_case[df_test_case['turn2']>0]
    else:
        df_test_case_tmp1 = df_test_case[(df_test_case['turn1']<1) | (df_test_case['turn2']<1)]
    
    tmp1_s = df_test_case_tmp1.sample(1)
    force_heal = tmp1_s['heal'].iloc[0]  # 最初でhealが決まるので取得しておく
    dicts = []
    more_sample_count = sample_count-1
    tmp1_rows = df_test_case_tmp1[(df_test_case_tmp1['heal'] == force_heal) & (df_test_case_tmp1.index != tmp1_s.index[0])] # healが共通なので、healでくくって取得する
    if len(tmp1_rows) > more_sample_count:
        tmp1_rows = df_test_case_tmp1[(df_test_case_tmp1['heal'] == force_heal) & (df_test_case_tmp1.index != tmp1_s.index[0])].sample(sample_count-1) # healが共通なので、healでくくって取得する

    df_test_case_tmp2 = pd.concat([tmp1_rows, tmp1_s], axis=0)
    for i, row1 in df_test_case_tmp2.iterrows():
        # print(row1)
        match row1['制約関数']:
            case 'finalHP_diamond':  equipment=["ダイアの鎧"]
            case 'finalHP_coral':    equipment=["珊瑚の指輪"]
            case 'finalHP_both':     equipment=["珊瑚の指輪", "ダイアの鎧"]

        hp_max, wave, thunder = row1['HP'], row1['wave'], row1['thunder']
        final_hp = calc_final_hp(hp_max, equipment, wave, thunder, force_heal)
        append_dict = {"hp_max":hp_max, "equipment":equipment, "wave":wave, "thunder":thunder, "heal":force_heal, "final_hp":final_hp}
        if (is_win_pattern and final_hp < 1) or ((not is_win_pattern) and final_hp > 0): # 負けたらおかしい or 勝ったらおかしいの組み合わせ
            print(f'is_win_pattern={is_win_pattern}')
            print(append_dict)
            raise Exception(append_dict)
            
        dicts.append(append_dict)

    df_result = pd.DataFrame(dicts)
    if len(dicts) < sample_count:  # 行が足りなかった時（負けパターンで多い）
        print(f'is_win_pattern={is_win_pattern} more_pattern')
        df_random = get_random_pattern(is_win_pattern, sample_count-len(dicts), force_heal)
        df_result = pd.concat([df_result,df_random], axis=0)
    
    return df_result
    
df_win_tmp1 = get_random_by_handmade(False,3)
df_win_tmp1
```

```python
# 勝利パターン / 敗北パターンを20回ずつ作成する
handmade_patterns = []
for is_win_pattern in [True,False]:
    for i in range(HANDMADE_WIN_LOSE_PATTERN):
        df_pattern1 = get_random_by_handmade(is_win_pattern,3)
        force_heal = df_pattern1['heal'].iloc[0]  # 最初でhealが決まるので取得しておく
        df_pattern2 = get_random_pattern(not is_win_pattern, 2, force_heal)
        df_pattern3 = pd.concat([df_pattern1, df_pattern2], axis=0)
        df_pattern4 = df_pattern3.sample(frac=1).reset_index(drop=True)
        handmade_patterns.append(df_pattern4)
```
