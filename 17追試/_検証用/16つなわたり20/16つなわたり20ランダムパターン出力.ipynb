# つなわたり10 ランダムパターン出力

```python
import os
import random
import pandas as pd

COUNT_OF_SAMPLES = 100

KEY_PATTERN = ["N","L","R","U","D"]
AIR_PATTERN = ["N","L","R","H","T"]
D_MIN = 10
D_MAX = 20
T_MAX = 30
```

```python
HANDMADE_TEST_CASE = [
    [['10 4', 'UUUU', 'TTTT'],['GOAL']],
    [['10 5', 'DUUUN', 'HTTTN'],['GOAL']],  # スタート地点より前には戻らない制約の確認（ゴールできるケース）
    [['10 7', 'DUULLLL', 'HTTTNHH'],['YOU LOST']],  # スタート地点より前には戻らない制約の確認（ゴールできないケース）
    [['10 7', 'DUULLLN', 'HTTTNHN'],['GOAL']],  # スタート地点より前には戻らない制約の確認（ゴールできるケース）
    [['10 8', 'DDUULLLN', 'HHTTTNHN'],['GOAL']],  # スタート地点より前には戻らない制約の確認（ゴールできるケース）
]
```

```python
def solve(input_list):
    # 入力を受け取る
    distance, total_time = map(int,input_list[0].split())
    keys = input_list[1]
    airs = input_list[2]
    
    # 左右の傾きを管理する変数
    tilt = 0  # 正: 右傾き、負: 左傾き
    time = 0
    pos = 0
    
    # 1文字ずつ処理
    for time in range(total_time):
        pos += 1
        sign = tilt//abs(tilt) if tilt!=0 else 0
        key = keys[time]
        air = airs[time]

        # キー入力
        match key:
            case 'L':
                tilt -= 1
            case 'R':
                tilt += 1
            case 'U':
                tilt += sign
                pos += 1
            case 'D':
                tilt -= sign
                pos -= 1
        
        # 風
        match air:
            case 'L':
                tilt -= 1
            case 'R':
                tilt += 1
            case 'H':
                pos -= 1
            case 'T':
                pos += 1

        pos = max(0,pos)
        print(f"time={time}, tilt={tilt}, pos={pos}, key={key}, air={air}, ")  # これは動作確認用のprint()文です。途中経過を見たい時にコメントアウトを外して、変数の内容を確認します。
        
        # 傾きチェック
        if abs(tilt) > 3:
            return ["YOU LOST"]
    else:
        # breakせずに最後まで到達できた場合
        return ["GOAL"]    
```

```python
for hand_made_test_case in HANDMADE_TEST_CASE:
    print(solve(hand_made_test_case[0]) == hand_made_test_case[1])
```

```python
def geenrateKeyAir (distance):
    pos = 0
    time = 0
    tilt = 0
    keys = ""
    airs = ""
    while pos < distance:
        key = random.choice(KEY_PATTERN)
        keys += key
        air = random.choice(AIR_PATTERN)
        airs += air
        sign = tilt//abs(tilt) if tilt!=0 else 0
        pos += 1
        time += 1

        # キー入力
        match key:
            case 'L':
                tilt -= 1
            case 'R':
                tilt += 1
            case 'U':
                tilt += sign
                pos += 1
            case 'D':
                tilt -= sign
                pos -= 1

        # 風
        match air:
            case 'L':
                tilt -= 1
            case 'R':
                tilt += 1
            case 'H':
                pos -= 1
            case 'T':
                pos += 1

        pos = max(0,pos)
        print(f"time={time}, tilt={tilt}, pos={pos}, key={key}, air={air}, ")  # これは動作確認用のprint()文です。途中経過を見たい時にコメントアウトを外して、変数の内容を確認します。
        
        # 傾きチェック
        if abs(tilt) > 3:
            #print(f"YOU LOST, tilt={tilt}, time={time}")  # これは動作確認用のprint()文です。途中経過を見たい時にコメントアウトを外して、変数の内容を確認します。
            result = False
            break
    else:
        result = True

    return result, distance, time, keys, airs

geenrateKeyAir(10)
```

```python
def generatePlayWidhGoal (req_distance, req_goal, not_list=[]):
    while True:
        is_goal, distance, time, keys, airs = geenrateKeyAir(req_distance)
        if is_goal != req_goal or time > T_MAX:
            continue

        inputs = ["{} {}".format(req_distance,time),keys,airs]
        output_array = solve(inputs)
        is_goal = (output_array[0] == "GOAL")
        if inputs in not_list:
            # print("重複", choice)
            continue

        print("d={}, goal={}, time={}, K={}, A={}".format(req_distance, is_goal, time, keys, airs))
        return inputs

generatePlayWidhGoal(10,True)
```

```python
def get_random_inputs(not_list=[]):
    while True:
        d_size = (D_MAX - D_MIN)+1
        answer_id = len(not_list)
        req_d = (answer_id % d_size) + D_MIN
        req_goal = (answer_id % (2*d_size)) >= d_size

        inputs = generatePlayWidhGoal(req_d, req_goal, not_list)
        
        return inputs
    raise ValueError('無限ループしたようです')
    
get_random_inputs()
```

```python
def output_all(out_xz_filename):
    all_inputs = []
    all_outputs = []

    for answer_id in range(COUNT_OF_SAMPLES):
        if answer_id < len(HANDMADE_TEST_CASE):  # 固定ケース
            all_inputs.append(HANDMADE_TEST_CASE[answer_id][0])
            all_outputs.append(HANDMADE_TEST_CASE[answer_id][1])
            if solve(HANDMADE_TEST_CASE[answer_id][0])[0] != HANDMADE_TEST_CASE[answer_id][1][0]:
                print(HANDMADE_TEST_CASE[answer_id][0])
                print(HANDMADE_TEST_CASE[answer_id][1])
                print(solve(HANDMADE_TEST_CASE[answer_id][0]))
                raise Exception('おかしい')
        else:
            all_inputs.append(get_random_inputs(all_inputs))
            all_outputs.append(solve(all_inputs[answer_id]))        # 解答を求める

    df_all = pd.DataFrame({"in":all_inputs, "out":all_outputs})
    df_all.index += 1
    os.makedirs(os.path.dirname(out_xz_filename), exist_ok=True)
    df_all.to_csv(out_xz_filename)

output_all("data/answer.csv.xz")
print('done')
```
